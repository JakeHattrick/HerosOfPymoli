```python
import pandas as pd
import numpy as np
```

```python
file = "Resources/purchase_data.csv"
dfm = pd.read_csv(file)
```

```python
dfm.head()
```

```python
playerCount = dfm['SN'].value_counts().size
playerCount
```

```python
uItem = dfm['Item Name'].value_counts().size
uItem
```

```python
totalP = dfm.count()
totalR = dfm["Price"].sum()
totalR
```

```python
maleDf = dfm[dfm['Gender']=='Male']
maleCount = maleDf['SN'].value_counts().size
maleCount
```

```python
femaleDf = dfm[dfm['Gender']=='Female']
femaleCount = femaleDf['SN'].value_counts().size
femaleCount
```

```python
otherDf = dfm[(dfm['Gender']!='Female')&(dfm['Gender']!='Male')]
otherCount = otherDf['SN'].value_counts().size
otherCount
```

```python
genDemo = pd.DataFrame([{"Gender":'Male',"Count":maleCount,
                         "Percentage":maleCount/playerCount},
                       {"Gender":'Female',"Count":femaleCount,
                         "Percentage":femaleCount/playerCount},
                       {"Gender":'Other',"Count":otherCount,
                         "Percentage":otherCount/playerCount},])
genDemo["Percentage"] = genDemo["Percentage"].map("{:.1%}".format)
genDemo.set_index('Gender',inplace = True)
```

```python
genDemo["Purchase Count"] = [len(maleDf),len(femaleDf),len(otherDf)]
genDemo["Average Purchase Price"] = [maleDf['Price'].mean(),
                                     femaleDf['Price'].mean(),
                                    otherDf['Price'].mean()]
genDemo["Average Purchase Price"] = genDemo["Average Purchase Price"].map("${:.2f}".format)
genDemo["Total Purchase Value"] = [maleDf['Price'].sum(),
                                   femaleDf['Price'].sum(),
                                   otherDf['Price'].sum()]
genDemo["Average Total per Person"] = genDemo["Total Purchase Value"] / genDemo["Count"]
genDemo["Total Purchase Value"] = genDemo["Total Purchase Value"].map("${:.2f}".format)
genDemo["Average Total per Person"] = genDemo["Average Total per Person"].map("${:.2f}".format)
```

```python
genDemo.set_index('Gender')
```

```python
ages = np.arange(dfm['Age'].min()-(dfm['Age'].min()%4),
                 dfm['Age'].max()+(dfm['Age'].max()%4),4)
ages
```

```python
labels = []
for i in range(len(ages)-1):
    labels.append(str(ages[i])+"-"+str(ages[i+1]))
labels[0] = "<"+labels[0].split('-')[1]
labels[len(labels)-1] = labels[len(labels)-1].split("-")[0]+"<"
labels
```

```python
dfm["Age Bracket"] = pd.cut(dfm['Age'],ages,labels = labels)
dfm.head()
```

```python
ageGroup = dfm.groupby('Age Bracket')
ageLists = []
for i in range(0,len(labels)):
    l = {}
    l.update({"Age Bracket":labels[i]})
    l.update({"Purchase Count":ageGroup['SN'].count()[i]})
    l.update({"Average Purchase Price":ageGroup['Price'].mean()[i]})
    l.update({"Total Purchase Value":ageGroup['Price'].sum()[i]})
    l.update({"Average Purchase per Person":ageGroup['SN'].count()[i] / len(ageGroup['SN'].unique()[i])})
    ageLists.append(l)
ageDem = pd.DataFrame(ageLists)
ageDem.set_index('Age Bracket')
#len(ageGroup['SN'].unique()[0])
```

```python
playerGroup = dfm.groupby("SN")
topPlayers = playerGroup["Price"].sum().sort_values(ascending = False)
topLists = []
for i in range(0,5):
    l = {}
    l.update({"SN":topPlayers.index[i]})
    l.update({"Purchase Count":playerGroup['Price'].count()[topPlayers.index[i]]})
    l.update({"Average Purchase Price":playerGroup['Price'].mean()[topPlayers.index[i]]})
    l.update({"Total Purchase Value":playerGroup['Price'].sum()[topPlayers.index[i]]})
    topLists.append(l)
topDf = pd.DataFrame(topLists)
topDf.set_index('SN')
```

```python
itemGroup = dfm.groupby("Item ID")
topItems = itemGroup["Price"].sum().sort_values(ascending = False)
topLists = []
for i in range(0,5):
    l = {}
    l.update({"ID":topItems.index[i]})
    l.update({"Name":itemGroup['Item Name'].value_counts().index[topItems.index[i]-1][1]})
    l.update({"Purchase Count":itemGroup['Price'].count()[topItems.index[i]]})
    l.update({"Item Price":itemGroup['Price'].mean()[topItems.index[i]]})
    l.update({"Total purchase Value":itemGroup['Price'].sum()[topItems.index[i]]})
    topLists.append(l)
topIDf = pd.DataFrame(topLists)
topIDf.set_index('ID')
#itemGroup["Item Name"].value_counts()
#topItems
```
